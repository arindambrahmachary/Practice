# Clipping and clapping method removing the outliers

Removing outliers from fAsym column

```python
# Define the percentile range for Winsorization
lower_percentile = 5  # Lower percentile
upper_percentile = 95  # Upper percentile


# Compute the lower and upper bounds for Winsorization
lower_bound = np.percentile(df['fAsym'], lower_percentile)
upper_bound = np.percentile(df['fAsym'], upper_percentile)

# Apply Winsorization to the variable with outliers
df['fAsym' + '_winsorized'] = df['fAsym'].clip(lower=lower_bound, upper=upper_bound)

# Check the effect of Winsorization
print("Before Winsorization:")
print(df['fAsym'].describe())

print("\nAfter Winsorization:")
print(df['fAsym' + '_winsorized'].describe())


sns.boxplot(df['fAsym' + '_winsorized'])


df['fAsym']=  df['fAsym' + '_winsorized']


```

#Data Transformation of our target variable

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()

df['class']= le.fit_transform(df['class'])
```

# Data partition

```python
x=df.iloc[:,0:10]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.2)
```

# model building with different model 

Desision Tree

```python
from sklearn.tree import  DecisionTreeClassifier
dc= DecisionTreeClassifier()
dc.fit(x_train,y_train)
y_pred_train= dc.predict(x_train)
y_pred_test= dc.predict(x_test)


from sklearn.metrics import accuracy_score
print("Training accuracy:", accuracy_score(y_train,y_pred_train))
print("Test accuracy:", accuracy_score(y_test,y_pred_test))
```

```python
#tree view

from sklearn import tree
#PLot the decision tree
tree.plot_tree(dc);
```

```python
#trying to observe max depth

max_depth= dc.tree_.max_depth
print("Max depth pf the tree: ", max_depth)
n_nodes = dc.tree_.node_count
print("Number of nodes in the tree:", n_nodes)

```

```python
#cross Validation 
training_accuracy=[]
test_accuracy= []


for i in range(1,101):
        dc=DecisionTreeClassifier(max_depth= 25)
        x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3, random_state=i)
        dc.fit(x_train,y_train)
        y_pred_train  = dc.predict(x_train)
        y_pred_test   = dc.predict(x_test)
        training_accuracy.append(accuracy_score(y_train,y_pred_train))
        test_accuracy.append(accuracy_score(y_test,y_pred_test))

print("Cross validation Training Accuracy: ",np.mean(training_accuracy).round(2))
print("Cross validation Test Accuracy: ",np.mean(test_accuracy).round(2))
```

#Random Forest

```python
from  sklearn.ensemble import RandomForestClassifier

rc= RandomForestClassifier(max_samples=0.5, n_estimators=100,max_features=0.7)
rc.fit(x_train,y_train)
y_pred_train= rc.predict(x_train)
y_pred_test= rc.predict(x_test)


from sklearn.metrics import accuracy_score
print("Training accuracy:", accuracy_score(y_train,y_pred_train))
print("Test accuracy:", accuracy_score(y_test,y_pred_test))
```

```python
# Random forest classifier with cross validation
training_accuracies = []
test_accuracies = []

for i in range(1,101):
        rc=RandomForestClassifier(max_depth= 25)
        x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3, random_state=i)
        rc.fit(x_train,y_train)
        y_pred_train  = rc.predict(x_train)
        y_pred_test   = rc.predict(x_test)
        training_accuracy.append(accuracy_score(y_train,y_pred_train))
        test_accuracy.append(accuracy_score(y_test,y_pred_test))

print("Cross validation Training Accuracy: ",np.mean(training_accuracy).round(2))
print("Cross validation Test Accuracy: ",np.mean(test_accuracy).round(2))
```
